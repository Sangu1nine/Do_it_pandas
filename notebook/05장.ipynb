```python
ebola = pd.read_csv('../data/country_timeseries.csv')
ebola
```

```python
ebola_long = ebola.melt(id_vars=['Date','Day'])
ebola_long
```

```python
var_split = ebola_long.variable.str.split('_')
var_split
```

```python
type(var_split)
```

```python
status_val = var_split.str.get(0)
country_val = var_split.str.get(1)
```

```python
ebola_long['status'] = status_val
ebola_long['country'] = country_val
ebola_long
```

```python
ebola_long = ebola.melt(id_vars=['Date','Day'])
```

```python
var_split = ebola_long.variable.str.split('_',expand = True)
var_split
```

```python
ebola_long[['status','country']] = var_split
ebola_long
```
